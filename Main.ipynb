## Run Actions randomly

```python
import numpy as np
from unityagents import UnityEnvironment

# Load the Unity environment
brain_names = env.brain_names  # Get all brain names
brains = {name: env.brains[name] for name in brain_names}  # Dictionary of brain objects

# Reset the environment to get environment info for all brains
env_info = env.reset(train_mode=False)

# Determine action sizes and the number of agents for each brain
action_sizes = {name: brain.vector_action_space_size for name, brain in brains.items()}
action_types = {name: brain.vector_action_space_type for name, brain in brains.items()}
num_agents = {name: len(env_info[name].agents) for name in brain_names}

for i in range(5):
    # Reset the environment for all brains, states, scores
    env_info = env.reset(train_mode=False)  
    states = {name: env_info[name].vector_observations for name in brain_names}  
    scores = {name: np.zeros(num_agents[name]) for name in brain_names}  

    while True:
        # get random actions
        actions = {}
        for name in brain_names:
            actions[name] = np.random.randint(action_sizes[name], size=num_agents[name])
        env_info = env.step(actions)

        # Update states, rewards, and scores for each brain
        for name in brain_names:
            next_states = env_info[name].vector_observations
            rewards = env_info[name].rewards
            dones = env_info[name].local_done
            scores[name] += rewards   
            states[name] = next_states  

        # Exit if any brain's agents are done
        if any(np.any(env_info[name].local_done) for name in brain_names):
            break
        
 
```

## Init hyperparameters

```python
GAMMA=0.99
LEARNING_RATE=1e-3
TAU=0.01
ALPHA=0.6
BETA=0.4
DEVICE = 'cuda'
BUFFER_SIZE = 1_000_000
BATCH_SIZE=128
ALPHA=0.7
```

## Start Training

```python
from DistributedAgent import DistributedDQN
import torch
from unityagents import UnityEnvironment
import numpy as np


env = UnityEnvironment(file_name="Soccer.x86_64")
brain_names = env.brain_names
brains = {name: env.brains[name] for name in brain_names}

env_info = env.reset(train_mode=True)
num_agents = {name: len(env_info[name].agents) for name in brain_names}

agents = DistributedDQN(
    DEVICE=DEVICE,
    env = env,
    brains=brains,
    num_agents=num_agents,
    gamma=GAMMA,
    lr=LEARNING_RATE,
    tau=TAU,
    alpha=ALPHA,
    beta=BETA,
    buffer_size=BUFFER_SIZE,
    batch_size=BATCH_SIZE
)

n_episodes = 1000
max_t = 1000
epsilon_start, epsilon_end, epsilon_decay = 1.0, 0.01, 0.995
epsilon = epsilon_start


for episode in range(1, n_episodes + 1):
    env_info = env.reset(train_mode=True)
    states = {name: env_info[name].vector_observations for name in brain_names}
    scores = {name: np.zeros(num_agents[name]) for name in brain_names}

    for t in range(max_t):
        actions = agents.act(states, epsilon)
        env_info = env.step(actions)

        next_states = {name: env_info[name].vector_observations for name in brain_names}
        rewards = {name: env_info[name].rewards for name in brain_names}
        dones = {name: env_info[name].local_done for name in brain_names}

        for name in brain_names:
            agent = agents.agents[name]
            for i in range(num_agents[name]):
                state = states[name][i]
                action = actions[name][i]
                reward = rewards[name][i]
                next_state = next_states[name][i]
                done = dones[name][i]

                td_error = reward + agents.gamma * np.max(agent['q_network'](torch.tensor(next_state, dtype=torch.float32).unsqueeze(0))) - agent['q_network'](torch.tensor(state, dtype=torch.float32).unsqueeze(0)).max().item()
                agent['replay_buffer'].add((state, action, reward, next_state, done), td_error)

        agents.learn()

        states = next_states
        for name in brain_names:
            scores[name] += rewards[name]

        if all(any(done) for done in dones.values()):
            break

    epsilon = max(epsilon_end, epsilon_decay * epsilon)
    print(f"Episode {episode} - Average Scores: {', '.join(f'{name}: {np.mean(scores[name]):.2f}' for name in brain_names)} - Epsilon: {epsilon:.2f}")

```
